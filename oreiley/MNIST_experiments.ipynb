# MNIST Dataset Import

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml("mnist_784", version = 1)
mnist.keys()
```

#### Data Stratification

```python
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(mnist["data"], 
                                                    mnist["target"], 
                                                    test_size=0.2, 
                                                    random_state=42)

```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

#int(x) for x in np.linspace(start = 1, stop = 15, num = 15)

param_1 = {
    "n_neighbors" : [5,10],
    "weights" : ["uniform", "distance"]
}

param_grid = [param_1]

knnclf = KNeighborsClassifier()
knnclf_grid_search = GridSearchCV(knnclf, param_grid, cv = 3, scoring = "accuracy", return_train_score = True,
                                  verbose = 2)
knnclf_grid_search.fit(X_train, y_train)
knnclf_grid_search.best_estimator_.get_params()
```
